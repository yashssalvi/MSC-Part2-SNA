## Aim: Write a program to compute the following for a given a network: (i) number of edges, (ii) number of nodes; (iii) degree of node; (iv) node with lowest degree; (v) the adjacency list; (vi) matrix of the graph.

```r
library(igraph)	
```

```r
g <- graph.formula(1-2, 1-3, 2-3, 2-4, 3-5, 4-5, 4-6,4-7, 5-6, 6-7)     
```

```r
plot(g)
```

### 1)no of edges

```r
ecount(g)
```

### 2)no of nodes

```r
vcount(g)
```

### 3)Degree Of nodes

```r
degree(g)
```

```r
dg <- graph.formula(1-+2, 1-+3, 2++3)
```

```r
plot(dg)
```

```r
degree(dg, mode="in")
```

```r
degree(dg, mode="out")
```

### 4) a) Node with lowest degree 

```r
V(dg)$name[degree(dg)==min(degree(dg))]
```

### 4) b) Node with lowest degree 

```r
V(dg)$name[degree(dg)==max(degree(dg))]
```

### 5) To find neighbours / adjacency list:

```r
neighbors(g,5)
```

```r
get.adjlist(dg)
```

### 6)Adjacency Matrix 

```r
get.adjacency(g)
```
